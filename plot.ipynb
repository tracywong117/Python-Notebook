# matplotlib.pyplot
## create figure and axes
Method1\
`plt.subplots()` returns a tuple containing a figure and axes object(s).\
`fig, ax = plt.subplots()` unpack the tuple into the variables `fig` and `ax`.\
`fig` is useful for changing figure-level attributes.\
axes objects have plotting methods
```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3) #rows:3, cols:1
ax[0].plot()
ax[1].plt()
ax[2].plt()
```

## customizing ticks
A formal way to customize ticks in Matplotlib is use:\
Locator class, and Formatter class.

### locator
Tick locators define the positions of the ticks.\
`import matplotlib.ticker as ticker`\
`axes_name.xaxis.set_major_locator(ticker.locator)`


### formatter
Tick formatters define how the numeric value associated with a tick on an axis is formatted as a string.
```python
import matplotlib.ticker as ticker
# Simple formatting
axs0[0].xaxis.set_major_formatter('{x} km')
axs0[1].xaxis.set_major_formatter(lambda x, pos: str(x-5))
# Null formatter
axs1[0].xaxis.set_major_formatter(ticker.NullFormatter())
# StrMethod formatter
axs1[1].xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.3f}"))

```


````python
#bar() function to be used with axes object is as follows − ax.bar(x, height, width, bottom, align)

import matplotlib.pyplot as plt
fig,ax= plt.subplots()
langs=['C', 'C++', 'Java', 'Python', 'PHP']
students=[23,17,35,29,12]
teachers=[30,48,30,40,20]
ax.bar(langs,students,label='number of students')
ax.bar(langs,teachers,bottom=students,label='number of teachers')

ax.set_ylabel('Number')
ax.set_xlabel('Programming language')
ax.set_title('Programming language choices')
ax.set_yticks(np.arange(0,100,20)) 
#plt.yticks(np.linspace(0,80,5)) 

ax.legend()
plt.show()
#plt.savefig('xxx.png')
````

````python
import numpy as np
import matplotlib.pyplot as plt

N=5

menMeans=(20, 35, 30, 35, 27)
womenMeans=(25, 32, 34, 20, 25)

ind=np.arange(N) # the x locations for the groups
width=0.35
fig,ax=plt.subplots()

ax.bar(ind,menMeans,width,color='r',label='Men')
ax.bar(ind,womenMeans,width,bottom=menMeans,color='b',label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

#ax.set_xticks, ax.set_yticklabels
ax.set_xticks(ind)
ax.set_xticklabels(['G1','G2','G3','G4','G5'])

ax.set_yticks(np.arange(0, 81, 10))
ax.legend()

plt.show()
````

````python
#twinx()

import matplotlib.pyplot as plt
import numpy as np

fig,a1=plt.subplots() 
x=np.arange(1,11)

a1.plot(x,np.exp(x),label='exp')
a1.set_ylabel('exp')

a2=a1.twinx()
a2.plot(x, np.log(x),'ro-',label='log')

a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
fig,ax=plt.subplots()

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

ax.hist(x, 50, density=True, facecolor='g', alpha=0.75)

ax.set_xlabel('Smarts')
ax.set_ylabel('Probability')
ax.set_title('Histogram of IQ')
ax.text(60,0.025,r'$\mu=100,\ \sigma=15$') # TeX equations
ax.axis([40, 160, 0, 0.03])
ax.grid(True)
plt.show()
````

````python
fig = plt.figure(figsize=(5, 12))
x = np.arange(0, 10, 0.1)
y = np.sin(x)
z = np.cos(x)
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)
ax1.set_title('Full view')
ax1.plot(y, color='blue', label='Sine wave')
ax1.plot(z, color='green', label='Cosine wave')
plt.ylim([-1, 1.8])
plt.xlim(xmin=10, xmax=80)
ax2.set_title('Partial view')
ax2.plot(y, color='blue', label='Sine wave')
ax2.plot(z, color='green', label='Cosine wave')
ax2.set_xlim([20, 60])
ax2.set_ylim(-1.2, 1.2)
ax3.set_title('Partial view')
ax3.plot(y, color='blue', label='Sine wave')
ax3.plot(z, color='green', label='Cosine wave')
ax3.legend(loc="upper right")
plt.ylim([-1, 1.8])
plt.xlim(xmin=10, xmax=80)
#Note: plt.xlim() and plt.ylim() affect the last axes, i.e. ax3. 
#Putting the calls before ax3.plot(…) has the same effect.
#So, axes_name.set_xlim() axes_name.set_xlim() should be used 
plt.show()
````
